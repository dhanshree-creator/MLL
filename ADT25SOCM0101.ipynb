**Step 1 Import Libraries**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler

print("Libraries imported successfully")
```

**Step 2: Load Dataset**

```python
df = pd.read_csv("heart.csv")
df.head()
```

```python
print(df.shape)
df.info()
```

**Step 3: Dataset Overview**

```python
print('Shape:', df.shape)
print('\nColumn Names:', df.columns.tolist())
df.head()
```

image.png

**Step 4: Data Types & Summary**

```python
print("\nData Types:\n")
print(df.dtypes)

print("\nStatistical Summary:\n")
df.describe(include="all")
```

image.png

**Step 5: Check Missing Values**

```python
print("\nMissing Values in Each Column:\n")
df.isnull().sum()
```

**Step 6: Handle Missing Values (if found)**

```python

# Example strategy: fill numeric columns with mean
for col in df.select_dtypes(include=['float64','int64']).columns:
   df[col].fillna(df[col].mean(), inplace=True)


# Fill categorical values with mode
for col in df.select_dtypes(include=['object']).columns:
   df[col].fillna(df[col].mode()[0], inplace=True)


df.isnull().sum()
```

**Step 7: Remove Duplicate Records**

```python
print("Duplicate Records Before:", df.duplicated().sum())

df.drop_duplicates(inplace=True)

print("Duplicate Records After:", df.duplicated().sum())
```

**Step 8: Encode Categorical Column**

```python
label_encoder = LabelEncoder()

categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    df[col] = label_encoder.fit_transform(df[col])

df.head()
```

image.png

**Step 9: Feature Scaling**

```python
scaler = StandardScaler()

scale_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

df[scale_cols] = scaler.fit_transform(df[scale_cols])

df.head()
```

image.png

** Step 10: Save Final Dataset**

```python
df.to_csv("final_heart_dataset.csv", index=False)
print("Final cleaned dataset saved as 'final_heart_dataset.csv'")
```



The Heart Disease dataset was successfully explored and preprocessed using standard EDA steps. Missing values and duplicates were handled, categorical features were encoded, and numerical features were scaled to ensure consistency. The final cleaned dataset is well-structured and ready for effective machine learning model development.
